```python
import tensorflow_datasets as tfds
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the facades dataset (without supervised labels)
dataset, info = tfds.load('cycle_gan/facades', with_info=True)

# Extract separate domains
trainA = dataset['trainA']
trainB = dataset['trainB']

# Preprocessing function (resizes and normalizes)
def preprocess(image):
    image = tf.image.resize(image, [256, 256])
    image = (tf.cast(image, tf.float32) / 127.5) - 1
    return image

# Apply preprocessing to each domain separately
trainA = trainA.map(lambda x: preprocess(x['image']))
trainB = trainB.map(lambda x: preprocess(x['image']))

# Combine them as (input, target) pairs
train_dataset = tf.data.Dataset.zip((trainA, trainB)).shuffle(1000).batch(1)

# Show a sample image pair
def show_images(input_img, target_img):
    plt.figure(figsize=(8, 4))
    titles = ['Input (Photo)', 'Target (Facade)']
    for i, img in enumerate([input_img, target_img]):
        plt.subplot(1, 2, i+1)
        plt.title(titles[i])
        plt.imshow((img + 1) / 2)  # convert [-1,1] -> [0,1]
        plt.axis('off')
    plt.show()

# Display one example
for input_img, target_img in train_dataset.take(1):
    show_images(input_img[0], target_img[0])

```
